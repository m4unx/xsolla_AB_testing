```python
#since country have too many variation, we not consider as variant. We only consider ui_version & experiment_group
```

```python
!pip install bayesian_testing
```

```python
from bayesian_testing.experiments import BinaryDataTest
```

```python
conv_test = BinaryDataTest()
```

```python
variant_A_desktop = dfm['convertion'][(dfm['experiment_group']=='A') & (dfm['ui_version']=='desktop')].values
variant_A_mobile = dfm['convertion'][(dfm['experiment_group']=='A') & (dfm['ui_version']=='mobile')].values
variant_B_desktop = dfm['convertion'][(dfm['experiment_group']=='B') & (dfm['ui_version']=='desktop')].values
variant_B_mobile = dfm['convertion'][(dfm['experiment_group']=='B') & (dfm['ui_version']=='mobile')].values
```

```python
conv_test.add_variant_data('A_desktop', variant_A_desktop)
conv_test.add_variant_data('A_mobile', variant_A_mobile)
conv_test.add_variant_data('B_desktop', variant_B_desktop)
conv_test.add_variant_data('B_mobile', variant_B_mobile)
```

```python
conv_test.probabs_of_being_best()
```

```python
#From the conversion point of view, variant A_desktop is the best with probability 99.98%
```
